Problem: Determine if a string of parentheses is a valid string.

Inputs: A string of only parentheses.

Outputs: Returns 'True' the parentheses are valid, returns 'False' if not.

Constraints: The input is 1 to 10000 characters long, and only contains brackets ('()[]{}')

Plan: Check for constraints, go through characters checking that the number of open and closed parenthesis is equal, and return whether they are.

## 2. Peer Review & Draft Code


Resources Reviewed: https://leetcode.com/problems/valid-parentheses/solutions/7108838/100-beats-0ms-stack-beginner-friendly-py-y4xp



```python
def validParentheses(s):
  unaccountedParentheses=0
  for char in s:
    if char in '([{': #Open parentheses
      unaccountedParentheses+=1
    else: #Closed parentheses
      if unaccountedParentheses==0: #If a closed parenthesis is there without any previous open parenthesis
        return False

      unaccountedParentheses=unaccountedParentheses-1

  return unaccountedParentheses==0

print(validParentheses("{}[]()"))
print(validParentheses("{}]()"))
```

```python
def isPowerOfThree(n,numToThree=0):
  if numToThree**3==n:
    return True
  if numToThree==n:
    return False

  return isPowerOfThree(n,numToThree+1)


print(isPowerOfThree(8))
```
